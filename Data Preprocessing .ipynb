```python
import warnings
warnings.filterwarnings("ignore", category = FutureWarning)

import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(src_dir, genres, song_samples):    
    # Empty array of dicts with the processed features from all files
    window =.1
    overlap =.5
    X = song_samples
    xshape = X
    print(xshape)
    chunk = int(xshape*window)# length of each one = 3 sec
    print(chunk)
    offset = int(chunk*(1.-overlap))#stride = 1.5 sec
    print(offset)
    hop_length = 512 # in num. of samples
    n_fft = 1024
    x= []
    y=[]

    # Read files from the folders
    
    for filenames in (os.listdir(src_dir)):
        if filenames.rfind(".") == -1 :
            print("\nProcessing: {}".format(filenames))
            for f in os.listdir("{}\\{}".format(src_dir,filenames)) :
            
                file_path ="{}\\{}\\{}".format(src_dir,filenames,f)
                signal, sr = librosa.load(file_path, sr=22000)
                
                
                if len(signal) < song_samples:
                    print('Error in {}\nNum of samples ( >= {}): {}'.format(file_path, song_samples, len(signal)))
                
                
                else:
                    signal = signal[:song_samples]
                    print("Reading file: {}".format(file_path))
                
                # Convert to dataset of spectograms/melspectograms
                    g = genres[filenames]
                    X = signal
                    for i in range(0, xshape-chunk + offset, offset):
## splitting song sample into parts 
                        p = X[i:i+chunk]
                        print(i,i+chunk)
                        k=librosa.feature.melspectrogram(p, n_fft = n_fft,hop_length = hop_length)[:,:,np.newaxis]
##converting amplitude signals to mel spectograms for each part
                        print(k.shape)
                       
                        x.append(k)
                        y.append(g)
                        print("genre no :" ,g)
                        print("no of mel spectograms till now" ,len(x))
    return np.array(x), np.array(y)  
```

```python
gtzan_dir = './genres/'
song_samples = 660000
genres = {'metal': 0, 'disco': 1, 'classical': 2, 'hiphop': 3, 'jazz': 4, 
          'country': 5, 'pop': 6, 'blues': 7, 'reggae': 8, 'rock': 9}
```

```python
  
X, y = read_data(gtzan_dir, genres, song_samples)
```
